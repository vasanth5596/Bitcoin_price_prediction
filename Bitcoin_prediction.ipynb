```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score
```

```python
df1=pd.read_csv('BTC_USD_2021-02-12_2022-02-11-CoinDesk.csv')
df1.head()
```

```python
df2=pd.read_csv('LTC_USD_2021-02-12_2022-02-11-CoinDesk (1).csv')
df2.head()
```

```python
sub_df=tot_df[tot_df['Currency']==bitcoin_type]
sub_df.head()
```

```python
sub_df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
def preprocess(df):
  
  data_scaled=scaler.fit_transform(df.iloc[:,2:3].values)
  cols=df.columns
  new_df=pd.DataFrame(data=data_scaled,columns=cols[2:3])
  # print(new_df.head())

  return new_df
```

```python
new_df=preprocess(sub_df)
new_df.head()
```

```python
train,test=new_df.iloc[0:int(len(new_df)*0.75),:],new_df.iloc[int(len(new_df)*0.75):,:]
print(f'length of train is {train.shape}, test size is {test.shape}')
```

```python
def split_data(data):
  X,y=[],[]
  for i in range(0,len(data),3):
    if i+3<len(data):
      X.append(data.iloc[i:i+2,:])
      y.append(data.iloc[i+2:i+3,:])
      # print(data.iloc[i:i+4,:]) # input
      # print(data.iloc[i+4:i+5,:]) # output
  return X,y
```

```python
X_train,y_train=split_data(train)
X_test,y_test=split_data(test)
print(f'length of X_train is {len(X_train)}, y_train is {len(y_train)}')
print(f'length of X_test is {len(X_test)}, y_test is {len(y_test)}')
```

```python
X_train=np.array(X_train)
X_test=np.array(X_test)
y_train=np.array(y_train)
y_test=np.array(y_test)
y_train=y_train.reshape(-1,1)
y_test=y_test.reshape(-1,1)
print(f'length of X_train is {(X_train.shape)}, y_train is {(y_train.shape)}')
print(f'length of X_test is {(X_test.shape)}, y_test is {(y_test.shape)}')
```

```python
from tensorflow.keras.layers import Dense,Input,LSTM,TimeDistributed,Bidirectional
from tensorflow.keras.models import Sequential


bi_lstm=Sequential()
bi_lstm.add(Input(shape=(2,1)))
bi_lstm.add(Bidirectional(LSTM(units=20,activation='relu')))
# lstm.add(LSTM(units=50,return_sequences=True))
bi_lstm.add((Dense(1)))

bi_lstm.compile(loss='mean_squared_error',optimizer='adam')

bi_lstm.summary()
```

```python
history=bi_lstm.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=20)
```

```python
y_pred_bi_lstm=bi_lstm.predict(X_test)
bi_lstm_acc=lstm_acc=(r2_score(y_test,y_pred_bi_lstm))
print("Accuracy of the model is ",bi_lstm_acc)
```

```python
plt.figure(figsize=[8,6])
plt.plot(history.history['loss'],'r',linewidth=2.0)
plt.plot(history.history['val_loss'],'b',linewidth=2.0)
plt.legend(['Training loss', 'Validation loss'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('loss',fontsize=16)
plt.title('val loss',fontsize=16)
```

```python
lstm=Sequential()
lstm.add(Input(shape=(2,1)))
lstm.add(LSTM(units=20,activation='relu'))
# lstm.add(LSTM(units=50,return_sequences=True))
lstm.add((Dense(1)))

lstm.compile(loss='mean_squared_error',optimizer='adam')

lstm.summary()
```

```python
history1=lstm.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=100)
```

```python
y_pred_lstm=lstm.predict(X_test)
lstm_acc=(r2_score(y_test,y_pred_lstm))
print("Accuracy of the model is ",lstm_acc)
```

```python
plt.figure(figsize=[8,6])
plt.plot(history1.history['loss'],'r',linewidth=2.0)
plt.plot(history1.history['val_loss'],'b',linewidth=2.0)
plt.legend(['Training loss', 'Validation loss'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('loss',fontsize=16)
plt.title('val loss',fontsize=16)
```

```python
from tensorflow.keras.layers import GRU
gru=Sequential()
gru.add(Input(shape=(2,1)))
gru.add(GRU(units=20,activation='relu'))
# lstm.add(LSTM(units=50,return_sequences=True))
gru.add((Dense(1)))

gru.compile(loss='mean_squared_error',optimizer='adam')

gru.summary()

```

```python
history2=gru.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=20)
```

```python
plt.figure(figsize=[8,6])
plt.plot(history2.history['loss'],'r',linewidth=2.0)
plt.plot(history2.history['val_loss'],'b',linewidth=2.0)
plt.legend(['Training loss', 'Validation loss'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('loss',fontsize=16)
plt.title('val loss',fontsize=16)
```

```python
y_pred_gru=lstm.predict(X_test)
gru_acc=(r2_score(y_test,y_pred_gru))
print("Accuracy of the model is ",gru_acc)
```
